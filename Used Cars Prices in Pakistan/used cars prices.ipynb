## 3. Outlier Detection and Removing Them
-	We visualize the outliers by making the boxplots. We made the box plots for the <strong>‘Price(in Rupees)’</strong>, <strong>‘Year’</strong>, <strong>‘Engine_cc’,</strong> and <strong>‘Mileage’</strong>. For each of these features outliers are detected.
-	IQR Method is used to remove the outliers.


```python
sns.boxplot(x=df['Price(in Rupees)'])
```

```python
sns.boxplot(x=df['Year'])
```

```python
sns.boxplot(x=df['Engine_cc'])
```

```python

sns.boxplot(x=df['Mileage'])
```

```python
Q1= df.quantile(0.25)
Q3= df.quantile(0.75)
IQR= Q3-Q1
print(IQR)
```

```python
df= df[~((df < ( Q1 - 1.5 * IQR )) | ( df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape

```
